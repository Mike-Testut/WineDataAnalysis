## Load Data

```python
import pandas as pd
```

```python
df = pd.read_csv("wines.csv")
```

## Wines with 100 Point rating

```python
wines_100pts = df.loc[df['points']==100]
```

```python
len(wines_100pts)
```

## Most Expensive Wine

```python
df.loc[df['price'] == df['price'].max()]['name'].squeeze()
```

## Ratings column from 0-5

```python
df["ratings"] = df["points"] / 20
```

```python
df["ratings"]
```

## Price Histogram under $100

```python
df.loc[df["price"] < 100]["price"].hist()
```

## Plot price by rating

```python
df.plot(x="price", y="points", figsize=(15,3), kind="scatter")
```
